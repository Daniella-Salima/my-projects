```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

### pairplot

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species', height=3)
plt.show()
```

### Boxplot

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=iris)
plt.show()
```

### violin plot

```python
species_colors = ['#78C851',  
                    '#F08030',  
                    '#6890F0',  
                  ]
                  
sns.violinplot( data=iris, palette=species_colors)
plt.show()
```

### The Histogram

```python
sns.histplot(iris, x='petal_length', kde=True)
penguins = sns.load_dataset("penguins")

sns.histplot(data=penguins, x="flipper_length_mm")
plt.show()
```

### Draw a Swarm Plot

```python
sns.swarmplot(y='petal_length', x='species', data=iris, palette=species_colors)
plt.show()
```

#### Bar Plot

```python
sns.barplot(penguins, x="island", y="body_mass_g")
plt.show()
```

### Heatmap Plot

```python
corrmat = iris.drop('species', axis=1).corr()
sns.heatmap(corrmat, annot = True, square = True)
plt.show()
```
