### Import MatPlotLib file

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

### The BARCHART

```python
x = np.array(["Apple", "Samsung", "Microsoft", "Acer"])
y = np.array([900, 700, 600, 300])
```

```python
plt.bar(x, y)
plt.show()
```

### The Histogram

```python
# Generate random data
data = np.random.randn(1000)

plt.hist(data, bins=30, density=True, color='pink', edgecolor='red', alpha=0.7)
plt.xlabel('M2')
plt.ylabel('Cheater')
plt.title('Cheater by m2')
plt.show()
```

### Pie Chart

```python
y = np.array([150, 520, 520, 480, 680])
plt.title('Chart')
plt.pie(y)
plt.show()
```
